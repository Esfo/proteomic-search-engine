```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx], idx

def scanfunc(scan):
    t = pd.DataFrame()
    timedict = {}
    t.loc[:,'m/z'] = scan['m/z array'].round(4)
    t.loc[:,'intensity'] = scan['intensity array']
    t.loc[:,'index'] = scan['index']
    timedict[scan['index']] = scan['scanList']['scan'][0]['scan start time']
    return t, timedict

def readfunc(f):
    msrun = mzml.MzML(f)
    ef = []
    times = {}
    for sc, td in msrun.map(lambda x: scanfunc(x)):
        ef.append(sc)
        times.update(td)
    return ef, times
```

```python
profiles = []
mt = time()
ef, times = readfunc(mzmlfile)
ef = pd.concat(ef)
currenttime = time() - mt
gc.collect()
print(currenttime, '- Time for tile extraction')
profiles.append(currenttime)
```

```python
mt = time()
et = ef.drop('intensity', axis=1)
et = et.groupby('m/z').count()

st = time()
et.plot.hist(bins=1000, figsize=(12,6))
plt.vlines(et.loc[(et > 1).to_numpy()].mean(), ymin=0, ymax=(et == et.mode().to_numpy()).count(), color='red')
plt.yscale('log')
plt.xlabel('Counts per unmerged mass channel')
plt.ylabel('Log Frequency')
plt.legend([])
plt.show()
nt = time() - st

et = et.loc[(et > et.loc[(et > 1).to_numpy()].mean()).to_numpy()]
currenttime = time() - mt - nt
print(et.loc[(et > 1).to_numpy()].mean()[0], '- Filtering Mean')
print(currenttime, '- Time for initializing filtering step')
profiles.append(currenttime)
```

The purpose here is to find centers of mass, which are masses that stick out with a high total intensity across the entire chromatogram. Essentially, if you were to sum the total signal intensity of every mass channel and look for the mass channels that stick out the most on a bar chart, these would be the centers of mass. This relies heavily on the fact that the MS data here is high-resolution data because I'm not binning any mass ranges at this point.

For the sake of removing noise, it's acceptable to remove datapoints that don't even hold an average number of counted indices with intensities greater than zero. This filtering will make finding the center of masses easier, and faster. These discarded channels can still come back into the game later when we merge mass channels.

Mass channels with a total signal count less than what is depicted above as the red line, the filtering mean, will not be present for the analysis of determining centers of mass.

```python
mt = time()
en = ef.set_index('m/z')
en.sort_index(inplace=True)
en = en.loc[et.index.array]
currenttime = time() - mt
print(currenttime, '- Time for applying filtering step')
profiles.append(currenttime)
```

```python
mt = time()
nbins = np.sqrt(en.index.unique().shape[0]).round().astype(int)
ediff = pd.DataFrame()
ediff.loc[:,'max'] = en.loc[:,'intensity'].max(level=0)
ediff.loc[:,'sum'] = en.loc[:,'intensity'].sum(level=0)
ediff.loc[:,'diff'] = ediff.loc[:,'sum'] - ediff.loc[:,'max']
ediff.loc[:,'dbins'] = pd.cut(ediff.loc[:,'diff'], bins=nbins, labels=False)

ediff.loc[:,'pdiff'] = ediff.loc[:,'diff'] / ediff.loc[:,'sum']
ediff.loc[:,'pdbins'] = pd.cut(ediff.loc[:,'pdiff'], bins=nbins, labels=False)

maxfreqloc = ediff.groupby('pdbins').count().sort_values('max').index.array[-1] #in case the mode function returns a list or whatever
#maxfreqloc = ediff.loc[:,'pdbins'].mode().to_numpy()[0]
maxbin = nbins - 1

filterloc = maxfreqloc - (maxbin - maxfreqloc)
fl = ediff.loc[ediff.loc[:,'pdbins'] == filterloc + 1, 'pdiff'].min()
vl = ediff.loc[ediff.loc[:,'pdbins'] == ediff.loc[:,'pdbins'].mode().to_numpy()[0], 'pdbins'].count()
currenttime = time() - mt
print(currenttime, '- Determining %Difference Cutoff')
profiles.append(currenttime)
```

```python
ediff.loc[:,'diff'].plot.hist(bins=nbins, figsize=(12,6))
plt.yscale('log')
plt.title('Distribution of Difference, $Sum - Maximum$, Intensities of Mass Channels')
plt.ylabel('Log Frequency')
plt.show()

fig, ax = plt.subplots(figsize=(12,6))
ediff.loc[:,'pdiff'].plot.hist(bins=nbins, ax=ax, alpha=0.5, label='Distribution of %Difference Intensity')
ediff.loc[ediff.loc[:,'dbins'] == 0, 'pdiff'].plot.hist(bins=nbins, ax=ax, alpha=0.5, label='Distribution of %Difference Intensity of the Most Frequent Difference Bin')
plt.vlines(fl, ymin=0, ymax=vl, color='black', label='Cutoff')
plt.legend()
plt.show()

ediff.plot.scatter(x='pdiff', y='sum', alpha=0.002, figsize=(12,6))
plt.vlines(fl, ymin=0, ymax=ediff.loc[:,'sum'].max(), color='black', label='Filter')
plt.yscale('log')
plt.ylabel('Sum of Intensities')
plt.xlabel('Percent Difference of $Sum - Maximum$ Intensities')
plt.show()
```

Mass channel intensities have been summed across their respective scans along the chromatogram. The difference between the maximum value for any channel, and the sum of that channel are noted above as their difference. Percent differences are noted as %difference.

Using the difference alone to determine a cutoff might induce bias towards higher-intensity peaks. Percent difference makes a more reasonable cutoff for keeping low-intensity peaks. This also keeps the large majority of data available for finding centers of mass, as shown in the second figure above.

The filterloc cutoff value assumes a symmetrical distribution of %difference of $sum - maximum$ intensity.

Summarized Reasoning:
 - If a mass channel's maximum value is close to it's sum, then there's not many other meaningful data points on that mass channel - meaning this could be off-target in terms of center of mass. It could still be within a useable ppm window of something that is a center, perhaps useful later. But for now, we're finding the center of mass.
 - If a mass channels's max is far away from it's sum, then it's clear there's many other data points - this could be a good starting indicator for a center of mass. These would be good data points to keep to find the true center of mass.

```python
mt = time()
masscenters = ediff.loc[ediff.loc[:,'pdbins'] > filterloc].index.array
ex = en.loc[masscenters]
es = pd.DataFrame()
es.loc[:,'intensity'] = ex.loc[:,'intensity'].sum(level=0)
es.loc[:,'index wmean'] = (ex.loc[:,'index'] * ex.loc[:,'intensity']).sum(level=0) / ex.loc[:,'intensity'].sum(level=0)
#above is the weighted mean of the found indices, or scan #, weighted by intensity of signal
es.loc[:,'index mean'] = ex.loc[:,'index'].mean(level=0)
es.loc[:,'index std'] = ex.loc[:,'index'].std(level=0)

firstderivmaxes = signal.argrelextrema(es.loc[:,'intensity'].to_numpy(), np.greater)[0]
secondderivmaxes = signal.argrelextrema(es.loc[:,'intensity'].to_numpy()[firstderivmaxes], np.greater)[0]
#thirdderivmaxes = signal.argrelextrema(en.to_numpy()[firstderivmaxes][secondderivmaxes], np.greater)[0]
currenttime = time() - mt
print(currenttime, '- Deriving Mass Centers')
profiles.append(currenttime)
```

Using the second derivative gave me less mass channels, but the channels had more peaks. Using only the first derivative here, the mass channels have much narrower +- ppm windows, and there are less peaks in each plotted spectra.

I chose to use only the first derivative in my analysis as the spectra were a bit cleaner. You can experiment for yourself to see which is more advantageous.

```python
nchannels = 1
deviation = 0.3
masschannel = 419.1
width = 0.4

an = 0
while True:
    fig, (a, ai) = plt.subplots(2,1, figsize=(16, 12))
    pinds = np.logical_and(es.index > masschannel, es.index < masschannel+width)
    etp  = es.loc[pinds]
    a.set_xlim(masschannel, masschannel+width)
    ai.set_xlim(masschannel, masschannel+width)
    
    a.vlines(es.index.array[firstderivmaxes], ymin=0, ymax=etp.loc[:,'intensity'].max(), linewidth=0.2, color='slateblue', label='First Derivative Local Maxes')
    a.vlines(es.index.array[firstderivmaxes][secondderivmaxes], ymin=0, ymax=etp.loc[:,'intensity'].max(), linewidth=0.2, color='indianred', label='Second Derivative Local Maxes')
    #a.vlines(en.index.array[firstderivmaxes][secondderivmaxes][thirdderivmaxes], ymin=0, ymax=etp.max(), alpha=0.8, color='green')
    
    etp.reset_index().plot.scatter(x='m/z', y='intensity', ax=a, s=0.1, color='purple')
    etp.reset_index().plot.scatter(x='m/z', y='index mean', ax=ai, s=0.1, color='purple', label='channel mean')
    etp.reset_index().plot.scatter(x='m/z', y='index wmean', ax=ai, s=0.1, color='green', label='intensity weighted channel mean')
    #etp.reset_index().plot.scatter(x='m/z', y='index mean', ax=ai, s=0.1, color='purple', yerr='index std', linewidth=0.05, alpha=0.4)
    a.set_yscale('log')
    ai.set_ylabel('index')
    masschannel += deviation
    an += 1
    aleg = a.legend()
    aileg = ai.legend()
    
    for lh in aleg.legendHandles:
        lh.set_alpha(1)
        lh.set_linewidth(1)
    
    for lh in aileg.legendHandles:
        lh.set_alpha(1)
        lh.set_sizes([1.5])
    
    plt.show()
    if an >= nchannels:
        break
```

If you wanted to find the cleanest mass channels, you could rank based on how well the average index matches the weighted index. The noise should offset the mean from the weighted mean, giving an indication of noise.

The groupings of the index values seem to deviate a bit in mass, and the mass channels still seem discernable. The centers of the masses become evident in the upper plots, where the sums of the mass channels themselves make peak-like shapes.

```python
mt = time()

fdmasses = es.index.to_numpy()[firstderivmaxes]
ppmdist = np.diff(fdmasses) / fdmasses[:-1] * 1000000
maxppmbridge = ppmdist.mean()

n = 0
lookahead = 5000
ecoords = []
while True:
    current = fdmasses[n:n+lookahead]
    ppmmatrix = np.abs(current.reshape(-1,1) - current) / current * 1000000
    bm = ppmmatrix < maxppmbridge
    diag = np.argwhere(bm[1:].diagonal() == False)
    zers = np.zeros(diag.shape).astype(int)
    zers[1:] += diag[:-1] + 1
    end = np.hstack((zers, diag)) + n

    if n + lookahead >= len(fdmasses):
        extra = np.array([diag.max() + 1, len(current) - 1]) + n
        end = np.vstack((end, extra))
        ecoords.extend(end.tolist())
        break

    ecoords.extend(end.tolist())
    n += diag.max() + 1

feout = fdmasses[ecoords]
mcppmwindows = np.abs((np.diff(feout)).flatten() / feout[:,1] * 1000000)
masschannels = feout.mean(axis=1)
mcppmwindows[mcppmwindows < maxppmbridge] = maxppmbridge
gc.collect()

currenttime = time() - mt

f, ax = plt.subplots(nrows=2, figsize=(16,12))
ax[0].hist(mcppmwindows, bins=np.ceil(np.sqrt(len(mcppmwindows))).astype(int))
ax[0].set_yscale('log')
ax[0].set_ylabel('Log Frequency')
ax[0].set_xlabel('+-ppm Window')

ax[1].hist(mcppmwindows, bins=np.ceil(np.sqrt(len(mcppmwindows))).astype(int))
ax[1].set_ylabel('Frequency')
ax[1].set_xlabel('+-ppm Window')

plt.show()

print('Mass Windows (+-ppm)')
print(f'Count: {len(mcppmwindows)}')
print(f'Min: {mcppmwindows.min()}')
print(f'Max: {mcppmwindows.max()}')
print(f'Mean: {mcppmwindows.mean()}')
print(f'Median: {np.median(mcppmwindows)}')


print(currenttime, '- Determining Mass Channels and Windows')
profiles.append(currenttime)
```

Farther below I'll show plots to indicate that when the mass channels are merged, there are a lot of examples where adjacent mass channels do not seem to reflect upon the same peaks - which is ideal. Otherwise you would be sampling the same peaks multiple times. This problem is probably not solvable by a single algorithm, as such there's likely no single perfect answer to prevent this from happening on every single mass channel. But filtering and deconvoluting data after peak widths are collected can help alleviate some of these effects later in the process.

```python
mt = time()

newcol = 'index'
newrow = 'm/z'
values = 'intensity'

#ef = ef.loc[ef.loc[:,'index'] < 5000] #If you're on a computer without a decent amount of RAM, < 16GB
#you could filter the amount of data here, to still get a representation of how the algorithm/analysis is working.

#def sparse_pivot(ef, newcol, newrow, values):
newcols = CategoricalDtype(sorted(ef.loc[:,newcol].unique()), ordered=True)
newinds = CategoricalDtype(sorted(ef.loc[:,newrow].unique()), ordered=True)
col = ef.loc[:,newcol].astype(newcols).cat.codes
row = ef.loc[:,newrow].astype(newinds).cat.codes
sm = sparse.csc_matrix((ef.loc[:,values], (row, col)), shape=(newinds.categories.size, newcols.categories.size))

#sm = sparse_pivot(ef, 'index', 'm/z', 'intensity')
efs = pd.DataFrame.sparse.from_spmatrix(sm, index=newinds.categories, columns=newcols.categories)

currenttime = time() - mt
print(currenttime, '- Organizing Mass Channels into Arrays')
profiles.append(currenttime)
```

```python
mt = time()
print('Assessing', sm.shape[0], 'channels for', len(masschannels), 'centers')

#assessing this number of arrays at once
magicnumber = 30000
endnum = np.floor((sm.shape[0] / magicnumber) - 1).astype(int)
mzindex = newinds.categories.to_numpy()

n = 0
v = 0
sms, mzs, mzws = [], [], []
start = magicnumber

while True:
    ipiece = sm[start-magicnumber:start]
    mpiece = mzindex[start-magicnumber:start]

    mci = np.logical_and(masschannels < mpiece.max(), masschannels > mpiece.min())
    mcp = masschannels[mci]
    mcw = mcppmwindows[mci]

    mppms = np.abs(mpiece.reshape(-1,1) - mcp) / mcp * 1000000
    boolmatrix = mppms < mcw
    boolmpiece = boolmatrix.any(axis=1)

    locations = np.argwhere(boolmatrix[boolmpiece])
    mcplocs = np.unique(locations[:,1], return_inverse=True)[1]

    newpiece = pd.DataFrame(ipiece[locations[:,0]].todense(), index=mcplocs).sum(level=0).to_numpy()
    mvals = np.unique(mcp[locations[:,1]]) 
    mzwvals = mcw[np.argwhere(mcp == mvals)].flatten()

    sms.append(sparse.csc_matrix(newpiece))
    mzs.append(mvals)
    mzws.append(mzwvals)
    v += newpiece.shape[0]
    sys.stdout.write(f'\r{n}/{endnum} {round(time() - mt, 4)}')
    sys.stdout.flush()
    n += 1
    start += magicnumber
    if start > sm.shape[0]:
        break

print(time() - mt)
gc.collect()

minpoints = et.min().to_numpy()[0]

sms = np.asarray(sparse.vstack((sms)).todense())
cinds = (sms > 0).sum(axis=1) > minpoints
#sms = sparse.csc_matrix(sms)
sms = sms[cinds]
mzs = np.asarray(list(itertools.chain(*mzs)))[cinds]
mzws = np.asarray(list(itertools.chain(*mzws)))[cinds]

medinds = np.median(sms, axis=1) == 0
sms = sms[medinds]
mzs = mzs[medinds]
mzws = mzws[medinds]

scanarray = np.arange(sms.shape[1])

currenttime = time() - mt
print(currenttime, '- Merging Mass Channels')
profiles.append(currenttime)
```

A great way to determine noisy channels is by determining which channels have a median greater than zero. Only ~120 of these were above zero-median on the file tested. You could see if any PSMs match up with these channels, it's doubtful but possible, and make a decent empirically-derived exclusion list.

In this analysis, I removed these channels from the analysis to focus on less-noisy data.

```python
start = 1
vals = []
while True:
    r, p = stats.pearsonr(sms[start-1], sms[start])
    vals.append(r)
    start += 1
    if start >= len(sms):
        break

vals = np.asarray(vals)

f, ax = plt.subplots(nrows=2, ncols=1, figsize=(16,12))

ax[0].hist(vals, bins=np.ceil(np.sqrt(len(vals))).astype(int))
ax[0].set_yscale('log')
ax[0].set_ylabel('Log Frequency')
ax[0].set_xlabel('Correlation Coefficients')

ax[1].hist(vals, bins=np.ceil(np.sqrt(len(vals))).astype(int))
ax[1].set_ylabel('Frequency')
ax[1].set_xlabel('Correlation Coefficients')

plt.show()
```

This is testing for correlation of neighboring mass channels without visualizing every single channel. A pearson coefficient is a sufficient comparator for this data because we're not looking to cluster these time-series, and it would be expected that a failed bridging of two related mass channels would show many of the same peaks in the same locations.

The large majority of directly adjacent channels do not correlate well with each other. There are clearly a few missed places, but that's not a huge deal.

```python
savgol = signal.savgol_filter(sms, window_length=101, polyorder=2, mode='wrap', deriv=0, axis=1)
savgol2 = signal.savgol_filter(sms, window_length=21, polyorder=0, mode='wrap', deriv=0, axis=1)

point = find_nearest(mzs, 420.32)
#point = find_nearest(mzs, 870.6)
#point = find_nearest(mzs, 616.53)
start = point[1] - 2 #indices prior to selected mass channel to start at
nchans = 3 #number of mass channels
channels = np.arange(nchans) + start

#limits for the peak plots, left-side
l = 0
h = sms.shape[1]
l = 6300 #left-hand limit for left plots
h = 6700 #right-hand limit for left plots

plt.rcParams.update({'font.size': 22})
for c in channels:
    f, a = plt.subplots(2,2, figsize=(20,12), sharex='col')
    n = mzs[c]
    w = n * (mzws[c] / 1000000)

    finds = np.logical_and(ef.loc[:,'m/z'] > n-w, ef.loc[:,'m/z'] < n+w)
    pinds = np.logical_and(es.index > n-w, es.index < n+w)
    fpi = np.logical_and(es.index[firstderivmaxes] > n-w, es.index[firstderivmaxes] < n+w)
    spi = np.logical_and(es.index[firstderivmaxes] > n-w, es.index[firstderivmaxes] < n+w)

    fp = ef.loc[finds]
    fp = fp.groupby('m/z').agg({'intensity': 'sum', 'index': 'mean'})

    etp  = es.loc[pinds]
    efd = es.index[firstderivmaxes][fpi]
    esd = es.index[firstderivmaxes][spi]

    a[0,0].plot(scanarray, sms[c], '-', color='orange', alpha=0.5)
    a[0,0].plot(scanarray, savgol[c], '-', color='green', alpha=0.5)
    a[0,0].plot(scanarray, savgol2[c], '-', color='brown', alpha=0.5)
    a[0,0].set_xlim(l, h)
    a[0,0].set_ylabel('intensity')
    #a.set_yscale('log')


    #fp.reset_index().plot.scatter(x='index', y='m/z', ax=a[1,0], s=0.1, color='green')
    #etp.reset_index().plot.scatter(x='index wmean', y='m/z', ax=a[1,0], s=0.2, color='purple')
    #a[1,0].set_xlim(l,h)


    fp.reset_index().plot.scatter(x='m/z', y='intensity', ax=a[0,1], s=0.5, color='green')
    etp.reset_index().plot.scatter(x='m/z', y='intensity', ax=a[0,1], s=0.5, color='purple')
    a[0,1].vlines(efd, ymin=0, ymax=etp.loc[:,'intensity'].max(), alpha=0.5, linewidth=0.3)
    a[0,1].vlines(esd, ymin=0, ymax=etp.loc[:,'intensity'].max(), alpha=0.08, color='red')
    a[0,1].set_yscale('log')
    a[0,1].yaxis.tick_right()
    a[0,1].yaxis.set_label_position("right")

    fp.reset_index().plot.scatter(x='m/z', y='index', ax=a[1,1], s=0.5, color='green')
    etp.reset_index().plot.scatter(x='m/z', y='index wmean', ax=a[1,1], s=0.5, color='purple')
    a[1,1].vlines(efd, ymin=0, ymax=etp.loc[:,'index wmean'].max(), alpha=0.5, linewidth=0.3)
    a[1,1].vlines(esd, ymin=0, ymax=etp.loc[:,'index wmean'].max(), alpha=0.08, color='red')
    a[1,1].yaxis.tick_right()
    a[1,1].yaxis.set_label_position("right")
    a[1,1].set_ylabel('index')

    widthbuffer = 20 #assumes peaks are at least 10 scans wide, which ain't too bad. Could probably also derive this number from the minimum scans per channels determined above - although it's softcoded. It might be wiser to keep this one hardcoded.
    maxis = signal.argrelmax(savgol[c])[0]
    minds = np.linspace(maxis-widthbuffer/2, maxis+widthbuffer/2, widthbuffer+1).transpose().astype(int)
    minds[minds >= len(savgol[c])] = len(savgol[c]) - 1
    minds[minds < 0] = 0
    maxif = maxis[savgol[c][minds].max(axis=1) == savgol[c][maxis]]
    maxin = maxif[savgol[c][maxif] > savgol[c][maxif].mean()]

    sgrad = np.gradient(savgol[c]) * 80 #scaling for the plot
    sgradsgd = signal.savgol_filter(sgrad, window_length=61, polyorder=2, deriv=1, mode='wrap') * 60
    sgdmif = signal.argrelmax(sgradsgd)[0]
    sinds = np.linspace(sgdmif-widthbuffer/2, sgdmif+widthbuffer/2, widthbuffer+1).transpose().astype(int)
    sinds[sinds >= len(sgrad)] = len(sgrad) - 1
    sinds[sinds < 0] = 0
    sgdm = sgdmif[sgradsgd[sinds].max(axis=1) == sgradsgd[sgdmif]]

    lm = np.ma.masked_array(np.repeat(sgdm.reshape(1,-1), len(maxin), axis=0), sgdm > maxin.reshape(-1,1))
    rm = np.ma.masked_array(np.repeat(sgdm.reshape(1,-1), len(maxin), axis=0), sgdm < maxin.reshape(-1,1))

    lpoints = np.asarray(lm.max(axis=1))
    rpoints = np.asarray(rm.min(axis=1))
    pcaps = np.hstack((lpoints.reshape(-1,1), rpoints.reshape(-1,1)))
    
    a[1,0].plot(scanarray, sms[c], '-', color='orange', alpha=0.5)
    a[1,0].plot(scanarray, savgol2[c], '-', color='brown', alpha=0.5)
    a[1,0].plot(scanarray, sgrad, '-', color='cyan', alpha=0.5)
    a[1,0].plot(scanarray, sgradsgd, '-', color='deeppink', alpha=0.5)
    a[1,0].plot(scanarray, savgol[c], '-', color='green', alpha=0.5)
    a[1,0].vlines(maxin, sms[c].max()*0.5, sms[c].max()*2, color='black', alpha=1)
    a[1,0].vlines(lpoints, 0, sgradsgd.max(), color='firebrick', alpha=0.5)
    a[1,0].vlines(rpoints, 0, sgradsgd.max(), color='navy', alpha=0.5)
    a[1,0].vlines(sgdm, 0, sms[c].max()*0.5, color='teal', alpha=0.5)
    a[1,0].set_xlabel('scan')
    for lc, rc in pcaps:
        a[1,0].hlines(sgradsgd.max(), lc, rc, color='black', alpha=1)
    a[1,0].set_xlim(l,h)
    a[1,0].set_ylabel('intensity')
    
    plt.suptitle(''.join((str(c), ': ', str(mzs[c]), ' +/-', str(mzws[c].round(4)), 'ppm')))
    f.subplots_adjust(hspace=0.05, wspace=0.05)
    plt.show()
```

This is a direct check of whether certain mass channels overlap. It's also a demonstration of some of the transforms used to find peaks, and their initial width measurements.

You can play with the parameters at the top of the cell to look at different mass channels, and to zoom in on different indices across the chromatogram for the left-hand plots.

To explain what is shown:
The plots on the right side are the same as what was shown above, the limits of the x-axis show the mass ranges being displayed as chromatograms on the left. The data shown is the sum of these mass channels.

The chromatograms on the left have 4 different transforms.
 - The green line is a savitsky-golay transformation used for finding local peak maximums. It's overly-rounded shape prevents from finding multiple maximums on the same peak.
 - The brown line is a savitsky-golay transformation with less rounding, and can be used to more accurately interpret peak widths at half-maximum.
 - The cyan line is the derivative of the green/rounded savitsky-golay transformation. It's used to derive the pink line.
 - The pink line is a savitsky-golay derived higher order derivative of the cyan line, that emphasizes where the lower points, or outer boundaries of the original savitsky-golay transform, approach the baseline. This gives a pretty accurate first impression of peak width at base height.

```python
mt = time()

widthbuffer = 20 #assumes peaks are at least 10, or this n/2, scans wide, which ain't too bad. Could probably also derive this number from the minimum scans per channels determined above - although it's softcoded. It might be wiser to keep this one hardcoded.
peakfill = 0.7
#This filters out things that seem to give on and off signal for whatever reason. Although even on major peaks I do see a lot of on and off signal, for reasons I can't explain. If it happens too much, I can kick it out via this.
prominencefilter = 5 #n times a prominence on at least one side
#maxis = signal.argrelmax(savgol, axis=1)
#sgrad = np.gradient(savgol, axis=1)
#sgradsgd = signal.savgol_filter(sgrad, window_length=61, polyorder=2, deriv=1, mode='wrap', axis=1)
#sgdmif = signal.argrelmax(sgradsgd, axis=1)

#use d = 5533 for testing, I guess
#d = 0 shows a good error where a single data point is transformed into a peak, this should be filtered out, due to there being no adjacency with this data.
#d = 479 for an example of the signal transforms failing to find a good peak base width, the need for corrections is there. 756?
enfs = []
for d in range(len(sms)):
    channel = sms[d]

    savgol = signal.savgol_filter(channel, window_length=101, polyorder=2, mode='wrap', deriv=0)
    savgol2 = signal.savgol_filter(channel, window_length=21, polyorder=0, mode='wrap', deriv=0)
    sgrad = np.gradient(savgol)
    maxis = signal.argrelmax(savgol)[0]
    sgd = signal.savgol_filter(sgrad, window_length=61, polyorder=2, deriv=1, mode='wrap')
    sdf = signal.argrelmax(sgd)[0]

    minds = np.linspace(maxis-widthbuffer/2, maxis+widthbuffer/2, widthbuffer+1).transpose().astype(int)
    minds[minds >= len(savgol)] = len(savgol) - 1
    minds[minds < 0] = 0
    maxif = maxis[savgol[minds].max(axis=1) == savgol[maxis]]
    maxin = maxif[savgol[maxif] > savgol[maxif].mean()]

    reminds = np.linspace(maxin-widthbuffer/2, maxin+widthbuffer/2, widthbuffer+1).transpose().astype(int)
    fininds = (channel[reminds] > np.median(channel)).sum(axis=1) >= widthbuffer * peakfill #removing peaks that are formed from a lone data point, it seems to be a common occurrence. Channels with median > 0 were already filtered in sms earlier.

    if any(fininds):
        maxfin = maxin[fininds]
        sinds = np.linspace(sdf-widthbuffer/2, sdf+widthbuffer/2, widthbuffer+1).transpose().astype(int)
        sinds[sinds >= len(sgrad)] = len(sgrad) - 1
        sinds[sinds < 0] = 0
        sgdm = sdf[sgd[sinds].max(axis=1) == sgd[sdf]]

        lm = np.ma.masked_array(np.repeat(sgdm.reshape(1,-1), len(maxfin), axis=0), sgdm > maxfin.reshape(-1,1))
        rm = np.ma.masked_array(np.repeat(sgdm.reshape(1,-1), len(maxfin), axis=0), sgdm < maxfin.reshape(-1,1))

        lpoints = np.asarray(lm.max(axis=1))
        rpoints = np.asarray(rm.min(axis=1))

        reminds = np.logical_or(lm.mask.all(axis=1), lm.mask.all(axis=1))
        if reminds.any():
            lpoints = lpoints[~reminds]
            rpoints = rpoints[~reminds]

        rpoints[rpoints > len(savgol)] = len(savgol) - 1

        #pcaps = np.hstack((lpoints.reshape(-1,1), rpoints.reshape(-1,1)))

        peaklocs = [] #center of where the peak is located

        rawheights = [] #maximum of the raw, untransformed data
        fitheights = [] #maximums of the savitsky-golay transformation
        rbasewidths = [] #the windows originally found via the third derivative transformation of the savitsky-golay transformation.
        tbasewidths = [] #windows that are trimmed using numpy's trim_zeros function, to determine if the rbasewidths is an overestimate of width at peak base
        hmbasewidths = [] #base widths extrapolated as assumed by a linear relationship between growth of the savitsky-golay transformation from it's maximum value to it's width at half-max and it's base. Essentially this is taken from extrapolating the base width to be 2x the distance of the width at half-max to the max from the max.
        halfheights = [] #height at half-max
        hmlbs = [] #left base according to hmbasewidths
        hmrbs = [] #right base according to hmabsewidths
        whms = [] #width at half-maxes, the width found across the distance of left and right half-max points, based on the savitsky-golay transformation.
        whmls = [] #the left-hand point used in calculating width at half-max.
        whmrs = [] #the right-hand point used in calculating width at half-max.
        whmlhs = [] #height of the left-hand point used in calculating width at half-max
        whmrhs = [] #height of the right-hand point used in calculating width at half-max
        peakrs = [] #R^2 values for correlation of savitsky-golay transformation and raw data
        peakpvals = [] #p-values for correlation of savitsky-golay transformations and raw data
        hmblmed = [] #median of values left of hmlb
        hmblmean = [] #mean of the values left of hmlb
        hmbfmed = [] #median of the values encompassed by hmbasewidths
        hmbfmean = [] #mean of the values encompassed by hmbasewidths
        hmbhmed = [] #median of the values encompassed by the half-max range
        hmbhmean = [] #mean of the values encompassed by the half-max range
        hmbrmed = [] #median of the values right of hmrb
        hmbrmean = [] #mean of the values right of hmrb

        rareas = [] #area under the savitsky-golay curve spanning the window found via third derivative of the savitsky-golay transformation
        hmbareas = [] #area under the savitsky-golay curve via the extrapolated bases taken from the widths at half-max
        hmareas = [] #area under the savitsky-golay curve within the half-max bounds

        iterinds = [] #index for keeping track of peaks

        lproms = [] #prominence of the left-most point of the original selection via third-derivative compared to the max of the savitsky-golay transform
        rproms = [] #prominence of the right-most point of the original selection via third-derivative compared to the max of the savitsky-golay transform
        endframe = pd.DataFrame()
        for m, l, r in zip(maxfin, lpoints, rpoints):
            lprominence = savgol2[m] / savgol2[l]
            rprominence = savgol2[m] / savgol2[r]
            if np.logical_and(lprominence > prominencefilter, rprominence > prominencefilter): #using logical_and filtered out a lot of one-sided, assymetrical peaks that were really just lop-sided blobs. This helps find peak-like shapes that are clearly resolved across the relevant mass channel.
                corr = stats.pearsonr(channel[l:r], savgol2[l:r])
                peakrs.append(corr[0])
                peakpvals.append(corr[1])

                peaklocs.append(m)

                rh = channel[l:r].argmax()
                rawheights.append(channel[l+rh])

                fh = savgol2[l:r].argmax()
                fitheights.append(savgol2[l+fh])

                rbasewidths.append(r - l)

                tbasewidths.append(len(np.trim_zeros(savgol2[l:r])))

                halfheightgoal = savgol2[l+fh]/2
                lhp = find_nearest(savgol2[l:l+fh], halfheightgoal)
                rhp = find_nearest(savgol2[l+fh:r], halfheightgoal)
                hh = np.mean([lhp[0], rhp[0]])

                lhmbw = l+fh-(fh-lhp[1])*2
                rhmbw = l+fh+rhp[1]*2

                hmbasewidths.append(rhmbw - lhmbw)
                halfheights.append(hh)
                hmlbs.append(lhmbw)
                hmrbs.append(rhmbw)

                lhmp = l+lhp[1]
                rhmp = l+fh+rhp[1]

                whms.append(fh + rhp[1] - lhp[1])
                whmls.append(lhmp)
                whmrs.append(rhmp)
                whmlhs.append(lhp[0])
                whmrhs.append(rhp[0])

                hmblmed.append(np.median(savgol2[l:lhmbw]))
                hmblmean.append(savgol2[l:lhmbw].mean())
                hmbfmed.append(np.median(savgol2[lhmbw:rhmbw]))
                hmbfmean.append(savgol2[lhmbw:rhmbw].mean())
                hmbhmed.append(np.median(savgol2[l+rhp[1]:l+fh+rhp[1]]))
                hmbhmean.append(savgol2[l+rhp[1]:l+fh+rhp[1]].mean())
                hmbrmed.append(np.median(savgol2[rhmbw:r]))
                hmbrmean.append(savgol2[rhmbw:r].mean())

                rareas.append(np.trapz(savgol2[l:r]))
                hmbareas.append(np.trapz(savgol2[lhmbw:rhmbw]))
                hmareas.append(np.trapz(savgol2[l+lhp[1]:r+fh+rhp[1]]))

                iterinds.append(d)

                lproms.append(lprominence)
                rproms.append(rprominence)

        if any(rawheights):
            endframe.loc[:,'peak location'] = peaklocs
            endframe.loc[:,'raw heights'] = rawheights
            endframe.loc[:,'fit heights'] = fitheights
            endframe.loc[:,'raw base widths'] = rbasewidths
            endframe.loc[:,'trimmed base widths'] = tbasewidths
            endframe.loc[:,'half-max based base widths'] = hmbasewidths
            endframe.loc[:,'height at half-max'] = halfheights
            endframe.loc[:,'left-hand half-max based base'] = hmlbs
            endframe.loc[:,'right-hand half-max based base'] = hmrbs
            endframe.loc[:,'width at half-max'] = whms
            endframe.loc[:,'left-hand half-max point'] = whmls
            endframe.loc[:,'right-hand half-max point'] = whmrs
            endframe.loc[:,'left-hand half-max height'] = whmlhs
            endframe.loc[:,'right-hand half-max height'] = whmrhs
            endframe.loc[:,'r-squared'] = peakrs
            endframe.loc[:,'p-value'] = peakpvals
            endframe.loc[:,'left-of-base median'] = hmblmed
            endframe.loc[:,'left-of-base mean'] = hmblmean
            endframe.loc[:,'right-of-base median'] = hmbrmed
            endframe.loc[:,'right-of-base mean'] = hmbrmean
            endframe.loc[:,'based-encompassed median'] = hmbfmed
            endframe.loc[:,'based-encompassed mean'] = hmbfmean
            endframe.loc[:,'width at half-max encompassed median'] = hmbhmed
            endframe.loc[:,'width at half-max encompassed mean'] = hmbhmean
            endframe.loc[:,'raw area'] = rareas
            endframe.loc[:,'base from width at half-max area'] = hmbareas
            endframe.loc[:,'width at half-max area'] = hmareas
            endframe.loc[:,'iteration number'] = iterinds
            endframe.loc[:,'left prominence'] = lproms
            endframe.loc[:,'right prominence'] = rproms
            endframe.loc[:,'mass channel'] = mzs[d]
            endframe.loc[:,'+- ppm'] = mzws[d]

            enfs.append(endframe)
enfs = pd.concat(enfs)
print(enfs.shape)
print(enfs.columns)

currenttime = time() - mt
print(currenttime, '- Assembling Peaks and Peak Properties')
profiles.append(currenttime)
```
